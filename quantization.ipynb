```python
import os
import cv2

# Paths
VAL_IMG_DIR = "/content/dataset/data/val"  # your validation images folder
OUTPUT_IMG_DIR = "calib_images"          # where to save ppm images
CALIB_TXT_PATH = "calibration.txt"       # list of images for NCNN

# YOLO input size
TARGET_SIZE = (640, 640)

# Create output folder
os.makedirs(OUTPUT_IMG_DIR, exist_ok=True)

# Collect image paths
img_files = [f for f in os.listdir(VAL_IMG_DIR) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

with open(CALIB_TXT_PATH, "w") as calib_file:
    for idx, img_name in enumerate(img_files):
        img_path = os.path.join(VAL_IMG_DIR, img_name)
        img = cv2.imread(img_path)
        if img is None:
            print(f"Skipping invalid image: {img_path}")
            continue

        # Resize to YOLO input size
        resized = cv2.resize(img, TARGET_SIZE)

        # Save as PPM (NCNN calibration requires uncompressed format)
        ppm_path = os.path.join(OUTPUT_IMG_DIR, f"{idx:05d}.ppm")
        cv2.imwrite(ppm_path, resized)

        # Write path to calibration file
        calib_file.write(ppm_path + "\n")

print(f"‚úÖ Calibration images saved to: {OUTPUT_IMG_DIR}")
print(f"‚úÖ Calibration file generated: {CALIB_TXT_PATH}")
```
